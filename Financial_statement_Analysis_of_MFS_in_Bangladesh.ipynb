The financial sector in Bangladesh is continuously growing in response to the evolving needs of the growing economy. Mobile financial services shortly known as MFS are a big contributor to the Bangladesh economy.
Since the inception of MFS in 2011, Bangladesh experienced a robust growth in number of account holders, volume and amount of transactions in MFS. A total of 13 mobile money services have now reached scale, with 1678207 agents, 803.93 Lac Active Accounts, 18497574 No daily average transaction and 3891.43 Core Average daily transaction [1]

MFS providing mainly 9 category products given as follows
- Inward Remittance
-  Cash In transaction
- Cash Out Transaction
-  P2P transaction
-  Salary Disbursement (B2P)
-  Utility Bill Payment (P2B)
-  Merchant Payment
-  Government Payment
-  Others

Bangladesh Bank website provides a comparative summary statement of every month. Downloaded the last 59 months (Dec-18 to Oct-23) data and after analysis of these summary statements

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

print('Reading CSV file ......')
# Read the CSV file into a DataFrame
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Mobile Financial Services (MFS).csv', low_memory=False, encoding='cp1252')

print(df.head(10))
```

```python
print(df.count())
```

```python
df['Date_info_New']=df['Date_info'].str[3:9]
print(df.head(10))
```

```python
data_1 = df.query('Description   == "No. of Banks currently providing the Services"')
plt.figure(figsize=(10, 6))
plt.bar(data_1['Date_info_New'], data_1['Data'], alpha=0.7, color='b', label='')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('Banks providing the Services')
plt.xticks(rotation=90);
```

```python
data_1 = df.query('Description   == "No. of agents"')
plt.figure(figsize=(10, 6))
plt.bar(data_1['Date_info_New'], data_1['Data'], alpha=0.7, color='b', label='')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('No. of agents')
plt.xticks(rotation=90);
```

```python
data_1 = df.query('Description   == "No. of registered clients in Lac"')
plt.figure(figsize=(10, 6))
plt.bar(data_1['Date_info_New'], data_1['Data'], alpha=0.7, color='b', label='')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('No. of registered clients in Lac')
plt.xticks(rotation=90);
```

```python
data_1 = df.query('Description   == "No. of active accounts in Lac*"')
plt.figure(figsize=(10, 6))
plt.bar(data_1['Date_info_New'], data_1['Data'], alpha=0.7, color='b', label='')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('No. of active accounts in Lac*')
plt.xticks(rotation=90);
```

```python
data_1 = df.query('Description   == "No. of total transaction"')
plt.figure(figsize=(10, 6))
plt.bar(data_1['Date_info_New'], data_1['Data'], alpha=0.7, color='b', label='')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('No. of total transaction')
plt.xticks(rotation=90);
```

```python
data_1 = df.query('Description   == "Total transaction in taka(in crore BDT)"')
plt.figure(figsize=(10, 6))
plt.bar(data_1['Date_info_New'], data_1['Data'], alpha=0.7, color='b', label='')
plt.xlabel('Date')
plt.ylabel('Number')
plt.title('Total transaction in taka(in crore BDT)')
plt.xticks(rotation=90);
```

```python
data_1 = df.query('Date_info_New == "Oct-23"')

data_1 = data_1.query('Description in ("Inward Remittance (crore BDT)","Cash In transaction (crore BDT)","Cash Out Transaction (crore BDT)","P2P transaction (crore BDT)","Salary Disbursement (B2P) (crore BDT)","Utility Bill Payment (P2B) (crore BDT)","Merchant Payment (crore BDT)","Government Payment (crore BDT)","Others")')

plt.figure(figsize=(10, 6))

bars = plt.bar(data_1['Description'], data_1['Data'], alpha=0.7, color='b', label='Transaction')

plt.xlabel('Description')
plt.ylabel('data')
plt.title('Transaction')
plt.xticks(rotation=45);

```
